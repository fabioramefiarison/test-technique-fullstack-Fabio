```

class Task:
    def __init__(self, name, project, priority, is_blocked=False, block_reason=None):
        self.name = name
        self.project = project
        self.priority = priority  # 1 = haute, 3 = basse
        self.is_blocked = is_blocked
        self.block_reason = block_reason

    def set_blocked(self, is_blocked: bool, block_reason: str | None = None):
        """
        Met à jour l'état de blocage de la tâche avec validation.
        """
        if is_blocked:
            if block_reason is None or block_reason.strip() == "":
                raise ValueError("Une raison valide doit être fournie lorsque la tâche est bloquée.")
            self.is_blocked = True
            self.block_reason = block_reason.strip()
        else:
            # Si on débloque la tâche, on supprime la raison
            self.is_blocked = False
            self.block_reason = None
```

```

def get_blocked_tasks(tasks: list[Task]) -> list[Task]:
    """
    Retourne uniquement les tâches bloquées,
    triées par priorité croissante (1 → 3),
    puis par nom (ordre alphabétique).
    """
    blocked_tasks = [task for task in tasks if task.is_blocked]

    return sorted(
        blocked_tasks,
        key=lambda task: (task.priority, task.name.lower())
    )
```

```

# Création de tâches
tasks = [
    Task("Configurer API", "Projet A", 1, True, "En attente des specs"),
    Task("Créer UI", "Projet A", 2),
    Task("Corriger bug login", "Projet B", 1, True, "Bug non reproductible"),
    Task("Documentation", "Projet A", 3, True, "Manque d'informations")
]

# Récupération des tâches bloquées triées
blocked = get_blocked_tasks(tasks)

print("Tâches bloquées triées :")
for task in blocked:
    print(f"- Priorité {task.priority} | {task.name} | Raison : {task.block_reason}")

# Test de validation
print("\nTest de validation :")
task_test = Task("Tests unitaires", "Projet C", 2)

try:
    task_test.set_blocked(True, "En attente d'environnement")
    print("Blocage valide appliqué avec succès.")
except ValueError as e:
    print("Erreur :", e)

try:
    task_test.set_blocked(True, "   ")
except ValueError as e:
    print("Erreur attendue :", e)
```
